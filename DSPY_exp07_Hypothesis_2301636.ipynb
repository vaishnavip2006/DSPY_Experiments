**What is need for Hypothesis Testing ? **


The **need for hypothesis testing** comes from the fact that we often want to make decisions or draw conclusions about a large population using only a sample of data.

### Why Hypothesis Testing is Needed

1. **Decision Making Under Uncertainty**
   In real-world situations, we don’t have complete information. Hypothesis testing helps us make logical decisions based on sample data instead of guesswork.

2. **To Test Assumptions (Claims)**
   It allows us to check whether a claim or assumption about a population is likely to be true or false.
   Example: “A new teaching method improves student performance.”

3. **To Validate Results Statistically**
   Instead of relying on intuition, hypothesis testing provides a scientific and mathematical basis to support conclusions.

4. **To Compare Groups or Methods**
   It helps determine whether differences between groups are significant or just due to random variation.
   Example: Comparing two medicines or two marketing strategies.

5. **To Control Errors**
   Hypothesis testing helps manage:

   * Type I Error (false positive)
   * Type II Error (false negative)

6. **Widely Used in Research and Industry**
   It is essential in fields like:

   * Business (A/B testing)
   * Medicine (clinical trials)
   * Engineering (quality control)
   * Social sciences (behavioral studies)

---

### In Short:

Hypothesis testing is needed to **make reliable, data-driven decisions and verify claims using statistical evidence rather than assumptions**.









**Q.2 With a neat diagram show important steps in hypothesis Testing.**

Here are the **important steps in Hypothesis Testing** along with a neat diagram:

---

## 📊 Steps in Hypothesis Testing

1. **State the Hypotheses**

   * Null Hypothesis (H₀): No effect / no difference
   * Alternative Hypothesis (H₁): There is an effect / difference

2. **Choose Significance Level (α)**

   * Common values: 0.05, 0.01
   * This defines the risk of rejecting H₀ when it is true

3. **Select Appropriate Test**

   * Based on data type and sample size
   * Examples: Z-test, t-test, Chi-square test

4. **Calculate Test Statistic**

   * Compute value using sample data

5. **Determine Critical Value / p-value**

   * Compare test statistic with critical value
   * Or use p-value approach

6. **Make Decision**

   * If p-value ≤ α → Reject H₀
   * If p-value > α → Fail to reject H₀

7. **Draw Conclusion**

   * Interpret the result in context of the problem

---

```
        Start
          │
          ▼
  State Hypotheses
   (H₀ and H₁)
          │
          ▼
 Choose Significance
     Level (α)
          │
          ▼
 Select Statistical Test
          │
          ▼
 Calculate Test Statistic
          │
          ▼
  Find p-value / Critical Value
          │
          ▼
   ┌───────────────────────┐
   │  Compare with α       │
   └───────────────────────┘
          │
   ┌──────┴──────┐
   ▼             ▼
p ≤ α         p > α
Reject H₀   Fail to Reject H₀
   │             │
   └──────┬──────┘
          ▼
   Draw Conclusion
          │
          ▼
         End
```

**Q.3 What is a student t distribution?**




The **Student’s t-distribution** is a probability distribution that is used in statistics when estimating the mean of a normally distributed population **in situations where the sample size is small** and the population standard deviation is unknown.

It was introduced by **William Sealy Gosset** under the pseudonym *Student*.

---

## Key Features of Student’s t-distribution

1. **Shape**

   * Similar to the standard normal distribution (bell-shaped and symmetric around 0)
   * Has **heavier tails**, meaning it is more prone to producing values far from its mean
   * As sample size increases, it approaches the standard normal distribution

2. **Parameters**

   * **Degrees of Freedom (df):**
     [
     df = n - 1
     ]
     where (n) is the sample size
   * The shape depends on the degrees of freedom: smaller df → heavier tails

3. **Usage**

   * Used for **small sample hypothesis tests**: t-test
   * Confidence intervals for the mean when population standard deviation is unknown

4. **Formula**
   For a sample mean (\bar{x}):
   [
   t = \frac{\bar{x} - \mu}{s / \sqrt{n}}
   ]
   where:

   * (\bar{x}) = sample mean
   * (\mu) = population mean (under H₀)
   * (s) = sample standard deviation
   * (n) = sample size

---

### ✅ In Short

The **Student’s t-distribution** is a **bell-shaped distribution used for small samples** when the population standard deviation is unknown, and it adjusts for extra uncertainty by having heavier tails than the normal distribution.

**Q.4 What is level of significance? What role t plays in hypothesis Testing**


Here’s a clear explanation of both concepts:

---

## **1. Level of Significance (α)**

The **level of significance**, denoted by **α**, is the **probability of making a Type I error**, i.e., rejecting the null hypothesis (H_0) when it is actually true.

* Common values: **0.05, 0.01, 0.10**
* It defines the **threshold for deciding whether the observed result is statistically significant**.
* Example:

  * If α = 0.05, there is a 5% risk of rejecting (H_0) incorrectly.

**Role in Hypothesis Testing:**

* Sets the **critical region** (rejection region) for the test statistic
* Helps in comparing the **p-value** with α to make a decision:

  * If **p-value ≤ α**, reject (H_0)
  * If **p-value > α**, fail to reject (H_0)

---

## **2. Role of t in Hypothesis Testing**

The **t-statistic** is used when:

* The population standard deviation is **unknown**, and
* The **sample size is small**

It measures how far the **sample mean** is from the **population mean** in terms of **standard error**:

[
t = \frac{\bar{x} - \mu}{s / \sqrt{n}}
]

Where:

* (\bar{x}) = sample mean
* (\mu) = population mean under (H_0)
* (s) = sample standard deviation
* (n) = sample size

**Role in Hypothesis Testing:**

1. The **t-statistic** is compared with **critical t-values** from the **Student’s t-distribution** table (based on α and degrees of freedom).
2. Determines whether to **reject or fail to reject (H_0)**.
3. Accounts for **extra uncertainty** when sample size is small, because the t-distribution has **heavier tails** than the normal distribution.

---

### ✅ **In Short**

* **Level of significance (α):** Threshold probability for making a Type I error.
* **t-statistic:** Measures how extreme the sample mean is and helps decide whether the result is statistically significant.

**Q.5 Design a hypothesis testing for identifying average age of students in TE IT class is 19 or not.**


 Let’s design a **complete hypothesis testing procedure** step by step for your problem: *“Is the average age of students in TE IT class 19 or not?”*

---

## **Step 1: State the Hypotheses**

We are testing whether the average age is 19.

* **Null Hypothesis (H₀):** The average age is 19
  [
  H_0: \mu = 19
  ]
* **Alternative Hypothesis (H₁):** The average age is not 19
  [
  H_1: \mu \neq 19
  ]

> This is a **two-tailed test** because we are checking “equal to or not equal to”.

---

## **Step 2: Choose Significance Level (α)**

* Typically, we take **α = 0.05** (5% risk of rejecting H₀ if it is true)

---

## **Step 3: Collect Sample Data**

* Suppose we take a **random sample of n students** from the class.
* Record their ages.

**Example:**

* Sample size (n = 25)
* Sample mean age (\bar{x} = 19.4)
* Sample standard deviation (s = 0.8)

---

## **Step 4: Select Appropriate Test**

* Population standard deviation is **unknown**, and the sample size is **small** → use **Student’s t-test**

---

## **Step 5: Calculate Test Statistic**

The **t-statistic** formula:

[
t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}}
]

Substitute the numbers:

[
t = \frac{19.4 - 19}{0.8 / \sqrt{25}} = \frac{0.4}{0.16} = 2.5
]

---

## **Step 6: Determine Critical Value**

* **Degrees of freedom (df) = n - 1 = 24**
* Two-tailed test, α = 0.05 → check t-table:

  * (t_{critical} \approx \pm 2.064)

---

## **Step 7: Make Decision**

* Compare **calculated t** with **critical t**:

  * (t_{calculated} = 2.5 > t_{critical} = 2.064)
* Since (|t_{calculated}| > t_{critical}) → **Reject H₀**

---

## **Step 8: Draw Conclusion**

* There is **sufficient evidence** to conclude that the **average age of students in TE IT class is not 19**.

---

### ✅ **Summary Table**

| Step         | Value / Action                    |
| ------------ | --------------------------------- |
| H₀           | μ = 19                            |
| H₁           | μ ≠ 19                            |
| α            | 0.05                              |
| Sample       | n = 25, (\bar{x} = 19.4), s = 0.8 |
| Test         | t-test (two-tailed)               |
| t-calculated | 2.5                               |
| t-critical   | ±2.064                            |
| Decision     | Reject H₀                         |
| Conclusion   | Average age is **not 19**         |

**Q.6 Check if you can accept H₀ at alpha 95%? Match your conclusions with python functions.**


 step by step.

We are testing:

* **H₀:** μ = 19
* **H₁:** μ ≠ 19
* Sample: n = 25, mean = 19.4, s = 0.8
* α = 0.05 → 95% confidence level

---

## **Step 1: Calculate t-statistic**

Formula:

[
t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}}
]

Substitute values:

[
t = \frac{19.4 - 19}{0.8 / \sqrt{25}} = \frac{0.4}{0.16} = 2.5
]

---

## **Step 2: Determine critical t-value**

* Degrees of freedom: df = n - 1 = 24
* Two-tailed test, α = 0.05 → critical t-value: ±2.064

---

## **Step 3: Compare t-statistic with critical t**

* |t-calculated| = 2.5
* t-critical = 2.064

Since **2.5 > 2.064**, we **reject H₀** at 95% confidence.

✅ Conclusion: **H₀ cannot be accepted at 95% confidence level**.

---

## **Step 4: Match with Python functions**

We can check the same using Python’s **scipy.stats**:

```python
from scipy import stats
import math

# Sample data
x_bar = 19.4
mu_0 = 19
s = 0.8
n = 25
alpha = 0.05

# t-statistic
t_stat = (x_bar - mu_0) / (s / math.sqrt(n))
df = n - 1

# Two-tailed p-value
p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))

print("t-statistic:", t_stat)
print("p-value:", p_value)

# Decision
if p_value < alpha:
    print("Reject H0")
else:
    print("Fail to reject H0")
```

**Expected Output:**

```
t-statistic: 2.5
p-value: 0.0195
Reject H0
```

* The **p-value (0.0195) < α (0.05)** → reject H₀
* Matches our manual calculation ✅
