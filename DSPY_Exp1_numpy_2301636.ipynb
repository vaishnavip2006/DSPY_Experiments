If any of these are unspecified, they default to the values start=0, stop=size of dimension, step=1. We'll take a look at accessing sub-arrays in one dimension and in multiple dimensions.

````python
x[::2]  # every other element
````

````python
x[2::2]  # every other element starting at index 2
````

**Introducing UFuncs**                                                               
For many types of operations, NumPy provides a convenient interface into just this kind of statically typed, compiled routine. This is known as a vectorized operation. This can be accomplished by simply performing an operation on the array, which will then be applied to each element. This vectorized approach is designed to push the loop into the compiled layer that underlies NumPy, leading to much faster execution.

````python
#Consider following loop based implementation to ind reciprocals for each element of an array
np.random.seed(0)

def compute_reciprocals(values):
    output = np.empty(len(values))
    for i in range(len(values)):
        output[i] = 1.0 / values[i]
    return output

big_array = np.random.randint(1, 100, size=1000000)
%timeit compute_reciprocals(big_array)
````

````python
#Same operatoin using UFuncs applying '/' over array elements
%timeit (1.0 / big_array)
````

Vectorized operations in NumPy are implemented via ufuncs, whose main purpose is to quickly execute repeated operations on values in NumPy arrays. UFuncs are extremely flexible – before we saw an operation between a scalar and an array, but we can also operate between two arrays as well as multidimentional arrays.

Computations using vectorization through ufuncs are nearly always more efficient than their counterpart implemented using Python loops, especially as the arrays grow in size. Any time you see such a loop in a Python script, you should consider whether it can be replaced with a vectorized expression.

** Q.1 are UFuncs in numpy**

UFuncs (universal functions) in NumPy are highly optimized, compiled C functions that perform element-wise operations on NumPy arrays. They enable fast, vectorized computations, which are significantly faster than using standard Python loops, especially for large datasets.

**Vectorization:** Ufuncs apply operations to entire arrays at once, pushing the looping into the compiled C layer, thus avoiding the overhead of Python's interpreter loop.                 

**Broadcasting:** They automatically support broadcasting, allowing operations on arrays of different shapes and sizes.

**Type Casting**: Ufuncs handle data type conversions internally to ensure operations are performed safely and efficiently.
Additional Methods: They provide useful methods such as reduce, accumulate, outer, and at for performing various aggregate and in-place operations.


C**ommon Types of NumPy UFuncs**                                                         
NumPy includes a wide variety of ufuncs, categorized by their functionality:

**Arithmetic Ufuncs**: These perform basic mathematical operations. Examples include np.add, np.subtract, np.multiply, np.divide, and np.sqrt. Python's standard arithmetic operators (like +, -, *) are often convenient wrappers around these ufuncs when working with arrays

**Trigonometric UFuncs:** These handle trigonometric calculations. Examples include np.sin, np.cos, np.tan, and their inverse functions like np.arcsin, which all operate on an element-by-element basis.

**Exponential and Logarithmic Ufuncs:**These functions work with exponents and logarithms. Common examples are np.exp, np.log, np.log2, and np.log10.
Bitwise Ufuncs: These perform operations at the bit level on integer data. They include functions like np.bitwise_and and np.bitwise_or

**Comparison/Logical Ufuncs**: These perform element-wise comparisons, returning boolean arrays. Examples include np.greater, np.equal, and np.logical_and.

Q.5 **Consider above 4×6 array, sample 2×3 sub array from left bottom of this array and store the result in variable named `subObjCpy`.**


To extract a **2×3** sub-array from the **left bottom** of a **4×6** array and store it in `subObjCpy`, you can use slicing with `.copy()` to create an independent array.

---

### ✅ Code Example

```python
import numpy as np

# Example 4×6 array (reshaped from 3×4×2)
Obj = np.random.randint(1, 51, size=(3, 4, 2))
Obj_reshaped = Obj.reshape(4, 6)

print("4×6 Array:\n", Obj_reshaped)

# Slice 2×3 sub-array from left bottom and store copy
subObjCpy = Obj_reshaped[2:4, 0:3].copy()

print("\n2×3 Left Bottom Sub-array (subObjCpy):\n", subObjCpy)
print("\nShape of subObjCpy:", subObjCpy.shape)
```

---

### ✅ Explanation

* `2:4` → selects the last 2 rows (bottom of the array).
* `0:3` → selects the first 3 columns (left side).
* `.copy()` → ensures `subObjCpy` is independent of the original array.

This gives you a **2×3 sub-array** from the **left bottom corner** of the 4×6 array

Q.7 **If you were to filter sub array view from above Obj object and want to apply modification in original and/or sub obj change respective element in other how will you achieve it?**

If you want modifications in the **sub-array** to **reflect in the original array**, you need to create a **view**, not a copy. In NumPy, slicing without `.copy()` automatically creates a **view**. Any changes made to the view will directly affect the original array, and vice versa.

Here’s how you can do it:

---

### ✅ Code Example Using a View

```python
import numpy as np

# Create original 3×4×2 array and reshape to 4×6
Obj = np.random.randint(1, 51, size=(3, 4, 2))
Obj_reshaped = Obj.reshape(4, 6)

# Extract 2×3 sub-array from left bottom as a view (no .copy())
subObjView = Obj_reshaped[2:4, 0:3]

print("Original 4×6 Array (Obj_reshaped):\n", Obj_reshaped)
print("\nSub-array view (subObjView):\n", subObjView)

# Modify an element in the view
subObjView[0, 0] = 999

print("\nSub-array view after modification:\n", subObjView)
print("\nOriginal 4×6 Array after modifying view:\n", Obj_reshaped)
```

---

### ✅ Explanation

1. `subObjView = Obj_reshaped[2:4, 0:3]` creates a view not a copy.
2. Changing `subObjView[0,0]` automatically updates `Obj_reshaped[2,0]`.
3. **Conclusion:**

   * A view shares memory with the original array.
   * Any modification in the view will directly affect the original array
   * Similarly, changes in the original array in that region will reflect in the view.
