## **Q.2 In IRIS data set which non graphical EDA is performed?**




In the Iris dataset, **non-graphical EDA (Exploratory Data Analysis)** means analyzing the data using **statistics and summaries instead of plots/graphs**.

Here are the main non-graphical EDA techniques performed on the Iris dataset 👇

---

## **1. Basic Structure Analysis**

* Number of **features (columns):** 4 (sepal length, sepal width, petal length, petal width)
* **Target variable:** species (Setosa, Versicolor, Virginica)
* **Number of tuples (rows):** 150

---

## **2. Data Types Check**

* All 4 features → **Numerical (continuous)**
* Target column → **Categorical (Nominal)**

---

## **3. Summary Statistics**

Using functions like `describe()`:

* **Mean**
* **Median**
* **Standard Deviation**
* **Minimum & Maximum**
* **Quartiles (Q1, Q2, Q3)**

👉 Example insights:

* Petal length has higher variance than sepal width
* Setosa species has smaller petal dimensions

---

## **4. Class Distribution**

* Setosa → 50 samples
* Versicolor → 50 samples
* Virginica → 50 samples

👉 Dataset is **perfectly balanced**

---

## **5. Missing Values Check**

* No missing/null values
* Clean dataset → no need for imputation

---

## **6. Duplicate Records**

* Check for duplicate rows
* Iris dataset generally has **no duplicates**

---

## **7. Correlation Analysis**

* Calculate correlation matrix (numerical relationships)

👉 Key findings:

* Petal length & petal width → **highly positively correlated**
* Sepal width → weak correlation with others

---

## **8. Group-wise Aggregation**

* Compute mean/median per species

👉 Example:

* Setosa → smallest petals
* Virginica → largest petals

---

## **9. Range & Spread Analysis**

* Range = max − min
* Helps identify variability in each feature

## **Conclusion**

Non-graphical EDA on the Iris dataset mainly involves:

* Statistical summaries
* Data type identification
* Missing value & duplicate checks
* Correlation analysis
* Class distribution

Q.3 ## **In Air Quality data set which graph is used to perform multivariate analysis?**


In the Air Quality dataset, **multivariate analysis** (i.e., analyzing relationships between multiple variables together) is commonly performed using the following graph:

---

## ✅ **Pair Plot (Scatter Plot Matrix)**

### **Why Pair Plot?**

* It shows **relationships between all pairs of variables**
* Helps identify:

  * Correlation between pollutants (e.g., CO, NO₂, O₃)
  * Patterns and clusters
  * Outliers

👉 Each cell in the matrix represents a **scatter plot between two variables**, while diagonals often show distributions.

---

## 🔹 Other Graphs Also Used for Multivariate Analysis

### **1. Heatmap (Correlation Matrix)**

* Displays correlation values between all variables
* Easy to identify strong positive/negative relationships

---

### **2. 3D Scatter Plot**

* Used when analyzing **3 variables at once**
* Example: CO vs Temperature vs Humidity

---

### **3. Line Plot (Multivariate Time series)
* Since air quality data is time-based
* Multiple variables plotted against time together

---

## ✅ **Final Answer (Short for Exam)**

**Pair Plot (Scatter Plot Matrix)** is the primary graph used for multivariate analysis in the Air Quality dataset because it shows relationships between multiple variables simultaneously.
