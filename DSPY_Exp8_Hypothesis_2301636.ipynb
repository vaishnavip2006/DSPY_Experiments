**What is 2 tailed and  tailed T Test ? **

. These are statistical tests used to determine whether there is a significant difference between the means of two groups. The distinction between **one-tailed** and **two-tailed** T-tests lies in the **direction of the hypothesis**.

---

## 1. **Two-Tailed T-Test**

* **Purpose:** Tests whether the sample mean is **significantly different** from the population mean (or another sample mean) **in either direction** — it doesn’t assume the difference has a specific direction.
* **Hypotheses:**

  * Null hypothesis ((H_0)): μ₁ = μ₂ (no difference)
  * Alternative hypothesis ((H_a)): μ₁ ≠ μ₂ (there is a difference)
* **Use case:** You just want to know if the means are different, **without specifying if one is higher or lower**.
* **Significance level:** Split across both tails of the distribution. For α = 0.05, each tail has 0.025.

**Example:**
You want to check if a new teaching method affects exam scores differently than the old method, but you don’t assume it will be higher or lower.

---

## 2. **One-Tailed (Single-Tailed) T-Test**

* **Purpose:** Tests whether the sample mean is **significantly greater than or less than** the population mean (or another sample mean) — a **specific direction** is assumed.
* **Hypotheses:**

  * Right-tailed: (H_0): μ₁ ≤ μ₂, (H_a): μ₁ > μ₂
  * Left-tailed: (H_0): μ₁ ≥ μ₂, (H_a): μ₁ < μ₂
* **Use case:** You are checking if one mean is specifically **larger or smaller** than the other.
* **Significance level:** Entire α is in **one tail**.

**Example:**
You want to test if a new fertilizer **increases** crop yield compared to the old fertilizer. You only care about improvement, not reduction.

---

### **Key Differences at a Glance**

| Feature                | Two-Tailed T-Test                       | One-Tailed T-Test                      |
| ---------------------- | --------------------------------------- | -------------------------------------- |
| Hypothesis Direction   | Checks for any difference (≠)           | Checks for specific direction (> or <) |
| α (Significance Level) | Split between both tails                | Entire α in one tail                   |
| When to Use            | Unsure of direction                     | Expect a specific direction            |
| Example                | Test if new drug changes blood pressure | Test if new drug lowers blood pressure |







**Q.2 Explain what is Type I and Type II errors with an example.**





break down **Type I** and **Type II errors** with examples so it’s clear.

---

## **1. Type I Error (False Positive)**

* **Definition:** This occurs when you **reject the null hypothesis** ((H_0)) even though it is **actually true**.
* **Think of it as:** Seeing an effect that **doesn’t exist**.
* **Probability:** Denoted by **α** (significance level, e.g., 0.05).

**Example:**
Suppose a new drug is being tested to see if it lowers blood pressure.

* Null hypothesis ((H_0)): The drug **does not lower blood pressure**.
* Alternative hypothesis ((H_a)): The drug **does lower blood pressure**.

If the drug **actually has no effect**, but the test says it does (we reject (H_0)), we’ve made a **Type I error**.

💡 Analogy: Like a fire alarm going off when there’s **no fire**.

---

## **2. Type II Error (False Negative)**

* **Definition:** This occurs when you **fail to reject the null hypothesis** ((H_0)) even though it is **actually false**.
* **Think of it as:** Missing an effect that **actually exists**.
* **Probability:** Denoted by **β**.

**Example:**
Using the same drug study:

* The drug **actually lowers blood pressure**, but your test fails to detect it (we don’t reject (H_0)).

This is a **Type II error**.

💡 Analogy: Like a fire alarm **not going off when there is a fire**.

---

### **Summary Table**

| Error Type  | What Happens                             | Analogy          | Symbol |
| ----------- | ---------------------------------------- | ---------------- | ------ |
| **Type I**  | Reject (H_0) when (H_0) is true          | False alarm      | α      |
| **Type II** | Fail to reject (H_0) when (H_0) is false | Missed detection | β      |

---

✅

* **Type I** = false positive → think “crying wolf.”
* **Type II** = false negative → think “missing the wolf.”

**Q.3 What is power of statistical test?**



---

## **Power of a Statistical Test**

* **Definition:**
  The **power** of a statistical test is the **probability that the test correctly rejects the null hypothesis ((H_0)) when the alternative hypothesis ((H_a)) is true**.

In other words:

[
\text{Power} = 1 - \text{Probability of Type II error} = 1 - \beta
]

Where:

* ( \beta ) = probability of **Type II error** (failing to detect an effect that exists)

* **Interpretation:** A higher power means the test is **more likely to detect a true effect**.

---

## **Key Points About Power**

1. **Range:** Power is between 0 and 1.

   * Power = 0.8 means there’s an 80% chance of detecting a true effect.

2. **Factors that Increase Power:**

   * **Larger sample size** → more information, less variability
   * **Larger effect size** → bigger difference between groups
   * **Higher significance level (α)** → easier to reject (H_0)
   * **Lower variability in data** → more precise measurement

3. **Why It Matters:**

   * Low power → high chance of Type II error (missing a real effect)
   * High power → confident detection of real differences

---

## **Example**

Suppose a new drug is believed to reduce blood pressure:

* (H_0): Drug has no effect
* (H_a): Drug reduces blood pressure

If the drug truly **reduces blood pressure**, the **power of the test** is the probability that your experiment will **detect this reduction** and reject (H_0).

* If power = 0.9 → 90% chance of detecting the drug’s effect
* If power = 0.5 → only 50% chance (not very reliable)





 **Q.4 Explain input and output characteristics for stats.ttest_rel function in your own words**



---

## **1. Purpose of `stats.ttest_rel`**

* `stats.ttest_rel` is used to perform a **paired (dependent) t-test**.
* **Paired t-test** checks whether the **mean difference** between two related samples is significantly different from zero.
* Example: Measuring blood pressure **before and after** a treatment on the same patients.

---

## **2. Input Characteristics**

`stats.ttest_rel(a, b, axis=0, nan_policy='propagate')`

| Parameter      | Description                                                                                                                                      |
| -------------- | ------------------------------------------------------------------------------------------------------------------------------------------------ |
| **a**          | First set of observations (array-like). Example: blood pressure **before** treatment.                                                            |
| **b**          | Second set of observations (array-like). Example: blood pressure **after** treatment. Must be same shape as `a`.                                 |
| **axis**       | The axis along which to perform the test. Default = 0 (along rows).                                                                              |
| **nan_policy** | How to handle NaN (missing) values: <br>• `'propagate'` → return NaN if NaN exists <br>• `'raise'` → throw an error <br>• `'omit'` → ignore NaNs |

**Important:**

* Both `a` and `b` must have the **same length** because each element in `a` is paired with a corresponding element in `b`.

---

## **3. Output Characteristics**

`stats.ttest_rel` returns **two values**:

1. **t-statistic (`t`)**

   * Measures the **size of the difference relative to the variability** in the differences.
   * Formula (simplified):

   [
   t = \frac{\text{mean}(a - b)}{\text{standard error of differences}}
   ]

   * **Positive t** → `a` is generally greater than `b`
   * **Negative t** → `a` is generally less than `b`

2. **p-value (`p`)**

   * Probability of observing such a t-statistic **if the null hypothesis is true**.
   * Small p-value (e.g., < 0.05) → reject (H_0) → difference is significant

---

### **Example**

```python
from scipy import stats

before = [120, 125, 130, 128, 122]
after = [118, 123, 129, 127, 121]

t_stat, p_value = stats.ttest_rel(before, after)
print(t_stat, p_value)
```

* **Interpretation:**

  * `t_stat` tells you the magnitude and direction of change
  * `p_value` tells you whether this change is statistically significant

---

✅ **Summary in Simple Words**

* **Input:** Two paired datasets of the same length (before/after or matched samples)
* **Output:**

  1. `t-statistic` → how big the difference is relative to variability
  2. `p-value` → whether the difference is statistically significant

**Q.5 What is function in stats module for identifying power of a statistical test. Explain output range of it and its significance.**



 **function in `stats` module for identifying the power of a statistical test**, its **output range**, and **significance**.

---

## **1. Function for Power Analysis**

In Python’s **SciPy**, the `stats` module itself doesn’t directly compute power. However, **`statsmodels.stats.power`** provides functions for power analysis. The most commonly used function is:

```python
from statsmodels.stats.power import TTestIndPower
```

* **`TTestIndPower().solve_power()`** can be used to calculate **power**, **effect size**, **sample size**, or **significance level** for a t-test.
* For example, for an **independent two-sample t-test**:

```python
from statsmodels.stats.power import TTestIndPower

analysis = TTestIndPower()
power = analysis.solve_power(effect_size=0.5, nobs1=30, alpha=0.05, ratio=1.0, alternative='two-sided')
print(power)
```

---

## **2. Output Range**

* **Power** is a **probability**, so its value is always between **0 and 1**.

[
0 \leq \text{Power} \leq 1
]

* **0** → test has no ability to detect the effect (useless)
* **1** → test will always detect the effect if it exists (perfect)
* Commonly, a power of **0.8 (80%)** is considered acceptable for most studies.

---

## **3. Significance of Power**

1. **Detecting True Effects:**

   * Power = 1 − β, where β is the probability of **Type II error**.
   * Higher power → lower chance of missing a real effect.

2. **Sample Size Planning:**

   * Helps determine how many samples you need to reliably detect an effect.

3. **Experimental Reliability:**

   * Low power → even if the effect exists, your test might fail to detect it.
   * High power → confident that a non-significant result means there’s likely **no real effect**.

---

### **Example Interpretation**

```python
power = 0.85
```

* **Interpretation:** There is an **85% chance** that the statistical test will correctly detect a true effect of the given size at the specified significance level.
* **Type II error probability (β)** = 1 − 0.85 = 0.15 → 15% chance of missing the effect.


Power analysis is essential **before conducting an experiment** to avoid wasting resources on a study that’s too small to detect meaningful effects.
